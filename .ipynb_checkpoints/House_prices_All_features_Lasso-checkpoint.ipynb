```python
df_train = df[:1460].copy()
target = 'SalePrice'
x = df_train[final_features].to_numpy()
y = df_train[target].to_numpy()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_norm = scaler.fit_transform(x)
x_norm.shape
```

```python
from sklearn.linear_model import LassoCV
from sklearn.model_selection import train_test_split
from sklearn.metrics import make_scorer, mean_squared_log_error
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_norm, y, test_size=0.25, random_state=17)
```

```python
Lasso = LassoCV()
```

```python
Lasso.fit(x_train, y_train)
```

```python
print(f'''
    
    Score:
    train {mean_squared_log_error(y_train, Lasso.predict(x_train))**.5 :.3f}
    test {mean_squared_log_error(y_test, Lasso.predict(x_test))**.5 :.3f}  

''')

pd.DataFrame(zip(final_features, Lasso.coef_)).sort_values(by=1, key=lambda r: -abs(r))
```

```python
Lasso.fit(x_norm, y)
```

```python
df_test = df[1460:].copy()
x_test = df_test[final_features].to_numpy()
x_test_norm = scaler.transform(x_test)
x_test.shape
```

```python
df_test['SalePrice'] = Lasso.predict(x_test_norm)
df_test['Id'] = df_test['Id'].astype('int32')
df_test[['Id', 'SalePrice']].to_csv('./house_Lasso.csv', index=False)
```
